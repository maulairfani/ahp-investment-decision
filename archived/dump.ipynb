```python
# libraries
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
 
```

```python

# Build a dataframe with your connections
df = pd.DataFrame({ 'from':['A', 'B', 'C','A','E','F','E','G','G','D','F'], 'to':['D', 'A', 'E','C','A','F','G','D','B','G','C']})
df
```

```python
df = pd.DataFrame({ 
    'from':['Investment Decision', 'Investment Decision', 'Investment Decision', 'Investment Decision', 'Sector', 'Sector', 'Sector'], 
    'to':['Sector', 'Index', 'Size', 'Quality', 'Teknologi', 'Kesehatan', 'Energi']
})
```

```python
# Build your graph
G=nx.from_pandas_edgelist(df, 'from', 'to')
 
# Spectral
nx.draw(G, with_labels=True, node_size=1500, node_color="skyblue", pos=nx.fruchterman_reingold_layout(G))
plt.title("spectral")
plt.show()
```
